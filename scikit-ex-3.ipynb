```python
from  sklearn import tree
from  sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('./dogs_n_cats.csv')
x_train = df[['Длина','Высота','Шерстист','Гавкает','Лазает по деревьям']]
y_train = df['Вид']
```

```python
df = pd.read_json('./dog_n_cats_test.json')
x_test = df[['Длина','Высота','Шерстист','Гавкает','Лазает по деревьям']]
```

```python
clf = DecisionTreeClassifier(criterion='entropy')
clf.fit(x_train, y_train)
```

```python
fn = ['Длина','Высота','Шерстист','Гавкает','Лазает по деревьям']
cn = ['собачка', 'котик']
tree.plot_tree(clf, filled=True, class_names=cn, feature_names=fn)
```
