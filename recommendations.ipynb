```python
movies = pd.read_csv('./movies.csv')
ratings = pd.read_csv('./ratings.csv')

movies.drop(['genres'], axis=1, inplace=True)
movies.head()
```

```python
ratings.drop(['timestamp'], axis=1, inplace=True)
ratings.head()
```

```python
user_item_matrix = ratings.pivot(index='movieId', columns='userId', values='rating')
user_item_matrix.head()
```

```python
user_item_matrix.fillna(0, inplace=True)
user_item_matrix.head()
```

```python
user_item_matrix.shape
```

```python
# отфильтровать активных пользователей (оставляли оценки чаще 20 раз)
# отфильтровать популярные фильмы (оценены чаще 10 раз)
```

```python
users_votes = ratings.groupby('userId')['rating'].agg('count')
movies_votes = ratings.groupby('movieId')['rating'].agg('count')

user_filter = users_votes[users_votes > 50].index
movie_filter = movies_votes[movies_votes > 10].index

user_item_matrix = user_item_matrix.loc[:,user_filter]
user_item_matrix = user_item_matrix.loc[movie_filter,:]
user_item_matrix.shape
```

```python
user_item_matrix.head()
```

```python
from scipy.sparse import csr_matrix
csr_data = csr_matrix(user_item_matrix.values)
print(csr_data[:2,:5])
```

```python
user_item_matrix = user_item_matrix.rename_axis(None, axis=1).reset_index()
user_item_matrix.head()
```

```python
from sklearn.neighbors import NearestNeighbors

knn = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=20, n_jobs=-1)

knn.fit(csr_data)
```

```python
search_word = 'Father'
recommendation = 10
# title, dist
movie_search = movies[movies['title'].str.contains(search_word)]
movie_search
```

```python
movie_id = movie_search.iloc[0]['movieId']
movie_id = user_item_matrix[user_item_matrix['movieId']==movie_id].index[0]
movie_id
```

```python
distances, indices = knn.kneighbors(csr_data[movie_id], n_neighbors=recommendation+1 )
```

```python
indices_list = indices.squeeze().tolist()
distances_list = distances.squeeze().tolist()

indices_distances = list(zip(indices_list, distances_list))
indices_distances[:5]
```

```python
indices_distances_sorted = sorted(indices_distances, key = lambda x: x[1], reverse=False)
indices_distances_sorted
```

```python
indices_distances_sorted = indices_distances_sorted[1:]
indices_distances_sorted
```

```python
for index_dist in indices_distances_sorted:
    movieId = user_item_matrix.iloc[index_dist[0]]['movieId']
    id = movies[movies['movieId'] == movieId].index
    title = movies.iloc[id]['title'].values[0]
    print(f'{title} - {index_dist[1]}')
```
